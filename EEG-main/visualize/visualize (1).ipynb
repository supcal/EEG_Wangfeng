```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 添加绘图框
fig=plt.figure(figsize=(10,10))

# 读取数据
data=pd.read_csv('/home/oliverhu/wangfeng/SSTG_re/visualize/data1.csv')

# 计算平均值和标准差
values = data.iloc[:, 1:].values

means = values[0,:]
stds = values[1,:]
max_value = values[2,:]
min_value = values[3,:]

# 绘制柱状图
plt.figure(figsize=(12,5))
plt.bar(data.columns[1:], means, yerr=stds, color=['#08509B'], label='Accuracy',linewidth=1,edgecolor='black',capsize=10,width=0.5)

# 在每个柱状图顶端显示对应的mean和std
for i, v in enumerate(means):
    plt.text(i-0.5, v + stds[i]+0.005, f'{v*100:.2f}({stds[i]*100:.2f})', fontsize=16)

# 设置纵坐标刻度大小和横坐标刻度大小
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

# 设置图例
plt.legend(loc='best', fontsize=18)
plt.grid(False)
# 设置纵坐标轴范围，横坐标轴名称，纵坐标轴名称
plt.ylim(0.75, 1)
plt.xlabel("The top-k ratio", fontsize=18)
plt.ylabel("Accuracy", fontsize=18)

# plt.show()
plt.savefig('/home/oliverhu/wangfeng/SSTG_re/visualize/kratio.pdf')
```

# DEAP sub 精度柱状图

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv('/home/wf/EEG_GTN/visualize/data_deap_dependent.csv')
print(data)
# 根据valence和arousal分类精度创建DataFrame
df = pd.DataFrame({
    'valence': data['valence'],
    'arousal': data['arousal']
})
df.index=data['序号']

# 画柱状图
ax = df.plot.bar(rot=0,width=0.6,figsize=(15,5),fontsize=16)
ax.set_xlabel('Subjects',fontsize=16)
ax.set_ylabel('Accuracy',fontsize=16)
plt.legend(loc='best', fontsize=18)
plt.grid(False)
# plt.savefig('/home/oliverhu/wangfeng/SSTG_re/visualize/deap_dependent.pdf')
plt.show()
```
